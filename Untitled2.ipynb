#REMOVING LAST 10 ROWS FOR MANUAL TESTING

```python
import pandas as pd # Ensure pandas is imported
df_fake_manual_testing = df_fake.tail(10)
df_fake.drop(df_fake_manual_testing.index, axis=0, inplace=True)

df_true_manual_testing = df_true.tail(10)
df_true.drop(df_true_manual_testing.index, axis=0, inplace=True)

df_fake.shape, df_true.shape
```

```python
data_merge = pd.concat([df_fake, df_true], axis =0 )
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data=data_merge.drop(["title", "subject","date"], axis = 1)
```

# CHECK FOR NULL VALUES WITHIN THE DATASETS

```python
data.isnull().sum()
```

```python
data= data.sample(frac = 1)
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(["index"], axis = 1, inplace = True)

data.columns
```

```python
data.head()
```

PROCESS THE TEXT

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

DEFINE THE DEPENDENT AND INDEPENDENT VARIABLES

```python
x=data['text']
y=data['class']
```

# SPLIT TRAIN TEST THE DATA

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size =0.25)
```

# NEXT IS TO CONVERT THE TEXT TO VECTORS

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

# NOW LET ME APPLY FIRST TRAINING MODEL - LOGISTICS REGRESSION

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

Predict the model and Accuracy score


```python
pred_lr=LR.predict(xv_test)
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

#NOW I WILL USE THE DECISION TREE MODEL TRAINING

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

#PREDICT THE MODEL

```python
pred_dt = DT.predict(xv_test)
```

#CHECK THE ACCURACY SCORE

```python
DT.score(xv_test, y_test)
```

# PRINT THE CLASSIFICATION REPORT FOR THE DECISION TREE MODEL

```python
print(classification_report(y_test, pred_dt))
```

# THEN I DO THE GRADIENT BOSSTING CLASSIFIER

```python
from sklearn.ensemble import GradientBoostingClassifier
GB = GradientBoostingClassifier(random_state=0)
GB.fit(xv_train, y_train)
```

#PREDICT THE MODEL

```python
pred_gb = GB.predict(xv_test)
```

THEN I CHECK ALSO THE ACCURACY SCORE

```python
GB.score(xv_test, y_test)
```

Print the Classification report

```python
print(classification_report(y_test, pred_gb))
```

# NOW I WILL DO THE RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(random_state=0)
RF.fit(xv_train, y_train)
```

Predict the model

```python
pred_rf=RF.predict(xv_test)
```

#CHECK THE ACCURACY SCORE OF RANDOM FOREST

```python
RF.score(xv_test, y_test)
```

Classification report

```python
print(classification_report(y_test, pred_rf))
```

NOW, I WILL WRITE TWO (2) CODES FOR THE MODEL TESTING

FIRST, THE - OUTPUT LABEL
SECOND, THE- MANUAL TESTING

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GBC = GB.predict(new_xv_test)
    pred_RFC = RF.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_lable(pred_LR[0]),                                                                                                       output_lable(pred_DT[0]),
                                                                                                              output_lable(pred_GBC[0]),
                                                                                                              output_lable(pred_RFC[0])))
```

TESTING ALL THE VARIOUS MODELS FOR "FAKE" AND "NOT FAKE" NEWS

```python
news = str(input())
manual_testing(news)
```
